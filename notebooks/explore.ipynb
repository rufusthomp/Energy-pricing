```python
import requests, time
import datetime as dt

url = 'https://data.elexon.co.uk/bmrs/api/v1/balancing/pricing/market-index'

start   = dt.datetime(2018, 1, 1) # MID data begins ~here
end     = dt.datetime.now() # up to today
current = start

all_records = [] # the accumulator across all windows

while current < end:
    window_end = current + dt.timedelta(days=7) # this window's far edge
    params = {
        'from': current.strftime('%Y-%m-%dT%H:%MZ'),
        'to':   window_end.strftime('%Y-%m-%dT%H:%MZ'),
    }
    r = requests.get(url, params=params)
    all_records.extend(r.json()['data'])
    print(current)
    current = window_end # advance the window
    time.sleep(0.1)                                   

price_df = pd.DataFrame(all_records)
```

```python
price_df['pv'] = price_df['price'] * price_df['volume']
grouped_price_df = price_df.groupby('startTime', as_index=False)[['pv', 'volume']].sum()
grouped_price_df['price'] = grouped_price_df['pv'] / grouped_price_df['volume']
grouped_price_df = grouped_price_df.dropna(subset=['price'])
grouped_price_df['startTime'] = pd.to_datetime(grouped_price_df['startTime']).dt.strftime('%Y-%m-%dT%H:%M:%S')

grouped_price_df
```
